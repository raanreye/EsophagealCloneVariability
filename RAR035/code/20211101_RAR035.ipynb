```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

# Test Plating

```python
Day1 = [1506,1226,1443,1326,1424,1184,]
Day3 = [3345,3785,2922,3406,3168,2974]

plate = [1506,1226,1443,1326,1424,1184,3345,3785,2922,3406,3168,2974]

plate = np.array(plate)
plate = plate.reshape(2,6)
plate = np.transpose(plate)
plate
```

```python

df = pd.DataFrame(plate, columns = ['300','1000'])

df.boxplot(column = ['300','1000']);
#plt.xlabel('Days',fontsize=15)
plt.ylabel('Number of cells in a well',fontsize=15)
plt.title('Plating for SplitBarcode Experiment',fontsize=18)  

# Save the default tick positions, so we can reset them...
locs, labels = plt.xticks() 
# Plot a line between the means of each dataset
plt.plot([1,2], [300*1.5**3,1000*1.5**3], 'or')
# Reset the xtick locations.
plt.xticks(locs)
plt.show()
```

```python
Day1 = [1127,1101,1176,1313]
Day3 = [3756,3942,3703,3463]

plate = [1127,1101,1176,1313,3756,3942,3703,3463]

plate = np.array(plate)
plate = plate.reshape(2,4)
plate = np.transpose(plate)
print(plate,np.mean(Day3)/5)
```
